```python
class ParcelaMango:
    """
    Representa una parcela individual con su sensor de humedad.
    Aplica limpieza de datos en el constructor (__init__).
    """
    def __init__(self, id_parcela, zona, humedad_cruda):
        self.id = id_parcela
        self.zona = zona
        
        # --- LÓGICA DE LIMPIEZA (POO) ---
        # Si la humedad es negativa, la ajustamos a 0. 
        # Si es mayor a 100, la ajustamos a 100.
        if humedad_cruda < 0:
            self.humedad = 0.0
        elif humedad_cruda > 100:
            self.humedad = 100.0
        else:
            self.humedad = float(humedad_cruda)
            
    def __repr__(self):
        return f"Parcela {self.id} en {self.zona}: {self.humedad}%"
    #--- logica de limpieza de datos en Zona ---
    @property
    def zona(self):
        return self._zona   
    @zona.setter
    def zona(self, valor):
        zonas_validas = ["Tepic", "Santiago", "San Blas"]
        if valor not in zonas_validas:
            print(f"Zona '{valor}' no es válida. Se asigna 'Desconocida'.")
            self._zona = "Desconocida"
        else:
            self._zona = valor
# Lista de datos  (Simulando un archivo externo)
lecturas_campo = [
    {"id": "PX1", "zona": "Tepic", "h": 65.5},
    {"id": "PX2", "zona": "Santiago", "h": 120.0}, # Error: Humedad > 100
    {"id": "PX3", "zona": "", "h": 58.2},
    {"id": "PX4", "zona": "Santiago", "h": -5.0},  # Error: Humedad negativa
    {"id": "PX5", "zona": "San Blas", "h": 90.0}
]
# Convertimos los dicts en OBJETOS de nuestra clase
objetos_parcelas = [
    ParcelaMango(obj['id'], obj['zona'], obj['h']) for obj in lecturas_campo
]

print(f"Se han creado {len(objetos_parcelas)} objetos validados.")

import pandas as pd

# Convertimos la lista de objetos a un DataFrame de Pandas
# Usamos un generador para extraer solo los atributos limpios
df_mangos = pd.DataFrame([
    {'ID': p.id, 'Zona': p.zona, 'Humedad': p.humedad} 
    for p in objetos_parcelas
])
print("\n--- DATAFRAME LISTO PARA ANÁLISIS ---")
print(df_mangos)
# --- ANÁLISIS ESTADÍSTICO ---
# Calculamos el promedio de humedad por Zona
# La fórmula que aplica Pandas internamente es:
# $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$
reporte_zonas = df_mangos.groupby('Zona')['Humedad'].mean()
print("\n --- HUMEDAD PROMEDIO POR ZONA ---")
print(reporte_zonas)
```

```python
import pandas as pd
import numpy as np
# 1. Carga de CSV
df_bio = pd.read_csv('biologia_marina.csv')
# 2. Creación de columna calculada (Vectorización)
df_bio['salinidad_ajustada'] = df_bio['salinidad_ppm'] * 1.05
# 3. Operaciones con NumPy
# Convertimos la columna a array para cálculos científicos
arr_temp = np.array(df_bio['temperatura'])
desviacion_std = np.std(arr_temp)
print(f"--- Análisis Biológico ---")
print(df_bio[['id_estanque', 'salinidad_ajustada']])
print(f"\nDesviación Estándar de Temperatura: {desviacion_std:.4f}")

from scipy import stats
# Prueba de Shapiro-Wilk para Normalidad
# Esta prueba evalúa si los datos provienen de una distribución normal
estadistico, p_valor = stats.shapiro(arr_temp)
print(f"--- Diagnóstico SciPy ---")
print(f"Estadístico W: {estadistico:.4f}")
print(f"P-Valor: {p_valor:.4f}")
# Interpretación Estadística (Nivel de significancia 0.05)
if p_valor > 0.05:
    print("Resultado: Los sensores están calibrados (distribución normal).")
else:
    print("Resultado: Anomalía detectada (los datos NO son normales).")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#creamos datos de ejemplo
data = {
    'metros_cuadrados': [60, 80, 100, 120, 150, 200, 250, 300, 50, 400],
    'precio_millones': [1.2, 1.8, 2.5, 2.1, 2.8, 4.5, 5.2, 5.8, 0.8, 8.5]
}
df = pd.DataFrame(data)
# Visualización inicial: ¿hay algo aqui?
sns.scatterplot(data=df, x='metros_cuadrados', y='precio_millones')
plt.title("Precio vs Metros Cuadrados (Datos Originales)")
plt.show()

x = df[['metros_cuadrados']]
y = df['precio_millones']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"Datos de entrenamiento: {len(x_train)}")
print(f"Datos para evaluar prueba: {len(x_test)}")    

#Entrenamiento e Inferencia
#1. Creamos el modelo
modelo = LinearRegression()
#2. Entrenamos el modelo
modelo.fit(x_train, y_train)
#3. Interpretacion
print(f"Beta_0 (Intercepto): {modelo.intercept_:.4f}")
print(f"Beta_1 (Pendiente): {modelo.coef_[0]:.4f}")
#4. Predicciones
#Queremos saber cuanto valdria una casa de 135 metros cuadrados
nueva_casa = np.array([[135]])
prediccion = modelo.predict(nueva_casa)
print(f"\nPredicción para casa de 135 m²: ${prediccion[0]:.2f} millones")
```

```python
import torch
from transformers import pipeline

device = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Usando dispositivo: {device}")

nlp = pipeline("sentiment-analysis", device=device)

textos = ["Me encanta el procesamiento del lenguaje Natural, es fascinante",
          "Esta materia me parece muy difícil y no entiendo nada",
          "El profesor muestra clases divertidas",
          "El examen estuvo mas o menos bien, aunque algunas preguntas fueron raras"]

for texto in textos:
    resultado = nlp(texto)[0]
    print(f"Texto: '{texto}'")
    print(f"Sentimiento: {resultado['label']} (Confianza: {resultado['score']:.4f})\n")
    print("-"*60)
```
