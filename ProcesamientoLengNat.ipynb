```python
import torch
from transformers import pipeline

device = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Usando dispositivo: {device}")

nlp = pipeline("sentiment-analysis", device=device)

textos = ["Me encanta el procesamiento del lenguaje Natural, es fascinante",
          "Esta materia me parece muy difícil y no entiendo nada",
          "El profesor muestra clases divertidas",
          "El examen estuvo mas o menos bien!, aunque algunas preguntas fueron raras"]

for texto in textos:
    resultado = nlp(texto)[0]
    print(f"Texto: '{texto}'")
    print(f"Sentimiento: {resultado['label']} (Confianza: {resultado['score']:.4f})\n")
    print("-"*60)
```

```python
import spacy
nlp = spacy.load("es_core_news_sm")

oraciones = ["El banco esta cerca del parque",
            "El banco me dio un préstamo",
            "El banco estaba ocupado",
            ]

for oracion in oraciones:
    doc = nlp(oracion)
    print(f"\nOración: '{oracion}'")
    print(f"Tokens: {[token.text for token in doc]}")
    entidades = [(ent.text, ent.label_) for ent in doc.ents]
    print(f"Entidades reconocidas: {entidades if entidades else 'Ninguna encontrada'}")
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
import nltk
nltk.download('punkt_tab')


texto = "¡Hola! ¿Cómo estás? Espero que estés teniendo un buen día niño. El procesamiento del lenguaje natural es increíble."

#TOKENIZAR

oraciones = sent_tokenize(texto)
print(f"Oraciones: {len(oraciones)}")

for i, oracion in enumerate(oraciones):
    print(f"Oración {i}: {oracion}")
    print()

tokens = word_tokenize(texto)
print(f"Tokens: {len(tokens)}")
print(f"Tokens: {tokens}")

from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

print(f"Total stopwords en español: {len(stopwords.words('spanish'))}")
print(f"Ejemplo de stopwords en español: {stopwords.words('spanish')[:313]}")

#filtramos tokens

tokens_filtrados = [token for token in tokens if token.lower() not in stopwords.words('spanish')]
print(f"Tokens después de filtrar stopwords: {len(tokens_filtrados)}")
print(f"Tokens filtrados: {tokens_filtrados}")
```

```python
# ── Celda 1: Imports y configuración general ────────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import seaborn as sns

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
from collections import Counter
import re
import string
import warnings

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams["figure.dpi"] = 110

print("Entorno listo.")
# → Entorno listo.


# ── Celda 2: Nuestro mini corpus ───────────────────────────────

corpus_raw = [
    # Positivas
    "Esta película es increíble y hermosa",
    "Me encanta esta película, es tan genial",
    "La actuación en esta película es maravillosa",
    "Una historia hermosa con una gran actuación",
    "Esta película es genial y me encanta",
    "Película maravillosa con una historia increíble",
    # Negativas
    "Esta película es terrible y aburrida",
    "Odio esta película, es tan mala",
    "La actuación en esta película es horrible",
    "Una historia aburrida con mala actuación",
    "Esta película es mala y la odio",
    "Película horrible con una historia terrible",
]

labels = ["pos", "pos", "pos", "pos", "pos", "pos",
          "neg", "neg", "neg", "neg", "neg", "neg" ]

print(f"Corpus cargado con {len(corpus_raw)} documentos.")
for i, doc in enumerate(corpus_raw):
    print(f"[{labels[i]}] {doc}")

# celda 3: Preprocesamiento básico ───────────────────────────────

def simple_preprocess(text):
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(f"[{re.escape(string.punctuation)}]", "", text)  # Eliminar puntuación
    tokens = text.split()  # Tokenizar por espacios
    return tokens

'''
#opcional: imprimir el resultado del preprocesamiento para cada documento
print("\nPreprocesando el corpus...")
corpus_preprocessed = [simple_preprocess(doc) for doc in corpus_raw]
for i, tokens in enumerate(corpus_preprocessed):
    print(f"[{labels[i]}] Tokens: {tokens}")
'''
corpus_tokenized = [simple_preprocess(doc) for doc in corpus_raw]

print(f"primeros 3 documentos tokenizados:")
for i, tokens in enumerate(corpus_tokenized[:3]):
    print(f"[{labels[i]}] Tokens: {tokens}")

#celda 4: Análisis de vocabulario ───────────────────────────────
all_tokens = [token for doc in corpus_tokenized for token in doc]
vocabulary = sorted(set(all_tokens),reverse=False)
word_to_idx = {word: idx for idx, word in enumerate(vocabulary)}

print(f"Tamaño del vocabulario: {len(vocabulary)}")

print(f"vocabulario completo: {vocabulary}")


# celda 5: vector de un documento ───────────────────────────────
def vectorize_document(tokens, word_to_idx):
    """convierte una lista de tokens en un vector de frecuencias basado en el vocabulario"""
    vector = np.zeros(len(word_to_idx), dtype=int)
    for token in tokens:
        if token in word_to_idx:
            idx = word_to_idx[token]
            vector[idx] += 1
    return vector

vec0 = vectorize_document(corpus_tokenized[0], word_to_idx)
print(f"Documento 0", corpus_raw[0])
print(f"Vector del documento 0: {vec0}")    

print(f"verificacion - palabras no-ceros en el vector:")
for word, idx in word_to_idx.items():
    if vec0[idx] > 0:
        print(f"'{word}': {vec0[idx]} veces")

#celda 6: Matriz Termino-Documento completa ───────────────────────────────
n_docs = len(corpus_tokenized)
n_terms = len(vocabulary)

bow_matrix_manual = np.zeros((n_docs, n_terms), dtype=int)
for i, tokens in enumerate(corpus_tokenized):
    bow_matrix_manual[i] = vectorize_document(tokens, word_to_idx)

print(f"Matriz BoW manual (shape: {bow_matrix_manual.shape}):")
print(bow_matrix_manual)


def_bow_manual = pd.DataFrame(bow_matrix_manual, 
                              columns=vocabulary, 
                              index=[f"doc_{i}" for i in range(n_docs)])

print(f"Matriz BoW manual como DataFrame:")
print(def_bow_manual)


# celda 7: Estadisticas básicas de la matriz ───────────────────────────────

total_cells = def_bow_manual.shape[0] * def_bow_manual.shape[1]
zero_cells = (def_bow_manual == 0).sum().sum()
sparsity = zero_cells / total_cells

print(f"Total de celdas: {total_cells}")
print(f"Celdas con cero: {zero_cells}")
print(f"Esparcidad: {sparsity}")
print(f"Celdas con información (no-cero): {total_cells - zero_cells}")

term_frequencies = def_bow_manual.sum(axis=0).sort_values(ascending=False)
print(f"Top 10 términos más frecuentes:")
print(term_frequencies.head(10))
```
