```python
class Alumno:
    def __init__(self, id, nombre, carrera, edad, genero, tel, correo):
        self.id = id
        self.nombre = nombre
        self.carrera = carrera
        self.edad = edad
        self.genero = genero
        self.tel = tel
        self.correo = correo
    def mostrar_reporte(self):
        print(f"ID: {self.id}, Nombre: {self.nombre}, Carrera: {self.carrera}, Edad: {self.edad}, Género: {self.genero}, Tel: {self.tel}, Correo: {self.correo}")   

alumno1 = Alumno(1, "Anzueto", "CD", 18, "Masculino", "555-1234", "anzueto@universidad.edu")
alumno2 = Alumno(2, "Zambrano", "CD", 20, "Masculino", "555-5678", "zambrano@universidad.edu")
alumno3 = Alumno(3, "Nieves", "CD", 24, "Masculino", "555-8765", "nieves@universidad.edu")

alumno1.mostrar_reporte()
alumno2.mostrar_reporte()
alumno3.mostrar_reporte()


```

```python
import pandas as pd

#--FASE 1: Definición de la clase LecturaEnergia --
class LecturaEnergia:
    """HOLA JAIME"""
    def __init__(self, id_medidor, consumo_kwh, pago):
        self.id_medidor = id_medidor
        self.consumo_kwh = consumo_kwh
        self.pago = pago
        self.categoria = "Alto" if consumo_kwh > 500 else "Medio" if consumo_kwh > 200 else "Bajo"
#creamos una lista de objetos POO puro
lista_objetos = [
    LecturaEnergia(10, 150,600),
    LecturaEnergia(2, 300,700),
    LecturaEnergia(3, 600,1000),
]

lecturas_de_campo= [{"id_medidor": 1, "consumo_kwh": 250, "pago": 800 },
                    {"id_medidor": 2, "consumo_kwh": 450, "pago": 900 },
                    {"id_medidor": 3, "consumo_kwh": 700, "pago": 1200 }]

#--FASE 2: Convertir la lista de objetos a un DataFrame de pandas --
datos_para_pandas = {
    "id_medidor": [obj.id_medidor for obj in lista_objetos],
    "consumo_kwh": [obj.consumo_kwh for obj in lista_objetos],
    "categoria": [obj.categoria for obj in lista_objetos],
    "pago": [obj.pago for obj in lista_objetos],
}

df_lecturas = pd.DataFrame(datos_para_pandas)
print("---Datos procesados en Pandas")
print(df_lecturas)  
print("consumo de promedio globlal:", df_lecturas["consumo_kwh"].mean())
print("pago promedio global:", df_lecturas["pago"].mean())

```
